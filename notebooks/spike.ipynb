```python
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
import librosa
import librosa.display
```

```python
model_path = "../pt/oiseaux.pt"
data = torch.load(model_path)
```

```python
data.squeeze()
```

```python
import matplotlib.pyplot as plt

plt.plot(data.squeeze())
```

```python
with open("../timit61.phoneset") as f:
    x_axis_labels = f.read().splitlines()
```

```python
fig, ax = plt.subplots(figsize=(50, 20))
sns.heatmap(data.squeeze().T, yticklabels=x_axis_labels, ax=ax)
```

```python
plt.figure(figsize=(14, 9))
audio_full = read("../audio/full-Sumi Jo-Caro mio ben.wav")
plt.figure(1)

plt.subplot(211)
plt.title('Spectrogram')
librosa.display.specshow(audio_full[1], x_axis='time', y_axis='log')

plt.subplot(212)
plt.title('Audioform')
librosa.display.waveshow(audio_full[1], sr=audio_full[0])
```
